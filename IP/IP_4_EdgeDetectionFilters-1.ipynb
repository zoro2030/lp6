# Sobel Filter
The Sobel filter is a widely used operator in image processing and computer vision for edge detection.

For a two-dimensional image, the Sobel filter approximates the gradient using convolution with two 3x3 kernels: one for detecting changes in the horizontal direction (often denoted as Gx) and another for changes in the vertical direction (often denoted as Gy).

```python
gx=cv2.Sobel(img,-1,1,0)
plt.imshow(cv2.cvtColor(gx,cv2.COLOR_BGR2RGB))
```

The parameters -1, 1, 0 specify that the Sobel operator should be applied along the horizontal axis (x direction). The -1 value for the depth (the third argument) indicates that the output image will have the same depth as the input image, preserving its data type.

```python
gy=cv2.Sobel(img,-1,0,1)
plt.imshow(cv2.cvtColor(gy,cv2.COLOR_BGR2RGB))
```

The parameters -1, 0, 1 specify that the Sobel operator should be applied along the vertical axis (y direction). The -1 value for the depth (the third argument) indicates that the output image will have the same depth as the input image, preserving its data type.

```python
# The parameters 0.5 and 0.5 indicate that the contribution of each image to the final result is equal.
gxgy=cv2.addWeighted(gx,0.5,gy,0.5,0)
plt.imshow(cv2.cvtColor(gxgy,cv2.COLOR_BGR2RGB))
```

# Laplacian Filter
The Laplacian operator is a second-order derivative filter used for edge detection and image sharpening. It computes the sum of second derivatives in a given neighborhood of each pixel.

```python
l=cv2.Laplacian(img,-1) #The -1 value indicates that the output image will have the same depth as the input image, preserving its data type
plt.imshow(l, cmap='gray')
plt.show()
```

# Canny edge
The parameters 50 and 150 are thresholds for the hysteresis procedure used in the Canny algorithm. Pixels with gradient values above the high threshold (150) are considered edges, while pixels with gradient values below the low threshold (50) are rejected unless they are connected to high threshold pixels. This hysteresis procedure helps suppress noise and detect true edges.



```python
c=cv2.Canny(img,50,150)
plt.imshow(cv2.cvtColor(c,cv2.COLOR_BGR2RGB))
```
