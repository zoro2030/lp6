```python
#Reading Image and showing it.
img = cv2.imread('/content/drive/MyDrive/Colab_Notebooks/IP/01/KatherinLangford.jpg')
plt.imshow(img)
```

```python
#converting BRG to RGB
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#The provided code resizes the image to a larger size of 1024x1024 pixels using OpenCV's cv2.resize() function.
i1=cv2.resize(img,(1024,1024))
plt.imshow(i1)
```

```python
i2=cv2.resize(img,(512,512))
plt.imshow(i2)
```

```python
#image of size 1024 is copied
copy=i1.copy()
plt.imshow(copy)
```

```python
# The original image size is 1024x1024 pixels, and pyrDown will reduce its size by half along each dimension
im1=cv2.pyrDown(copy)
plt.imshow(im1)
```

```python
#This code is repeatedly applying the pyrDown function from OpenCV to further reduce the resolution of the image.
for i in range(4):
    im1=cv2.pyrDown(im1)
    plt.imshow(im1)
```
