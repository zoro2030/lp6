```python
# Display the grayscale image flag 0 indicates that.
e = cv2.imread('/content/drive/MyDrive/Colab_Notebooks/IP/01/KatherinLangford.jpg', 0)
plt.imshow(e, cmap='gray')
```

Contrast Stretching:

```python
original = e.copy() # Copy the original grayscale image
xp = [0, 64, 128, 192, 255] # Define input and output intensity levels for histogram equalization
fp = [0, 16, 128, 240, 255]
x = np.arange(256)  # Generate lookup table using linear interpolation
table = np.interp(x, xp, fp).astype('uint8')
p = cv2.LUT(e, table) # Apply histogram equalization using the lookup table
plt.imshow(p, cmap='gray') # Display the contrast image
plt.show()
```

Grayscale Slicing:

```python
min_range = 0 # Specify the min and max range
max_range = 50

mask = (e > min_range) & (e < max_range) # Create a mask for pixels within the desired range

img1 = np.where(mask, 255, p) # Create a new image with pixels within the range set to 255

# Display the resulting image
plt.imshow(img1, cmap='gray')
plt.show()
```
