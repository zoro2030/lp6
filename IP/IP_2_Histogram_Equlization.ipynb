```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
image=cv2.imread('/content/drive/MyDrive/Colab_Notebooks/IP/01/KatherinLangford.jpg')
```

```python
plt.imshow(image)
plt.title("BRG Image")
image.shape
```

```python
# Convert BGR to RGB
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.title("RGB Image")
plt.imshow(img_rgb)
```

```python
# Convert BGR to Gray Sacle
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.title("Gray Image")
plt.imshow(gray_img)
```

```python
plt.hist(img_rgb.ravel(),256, [0,256])
plt.title("Original Image Histogram")
plt.xlabel("Gray levels")
plt.ylabel("No of Pixels")
```

```python
plt.hist(gray_img.ravel(),256, [0,256])
plt.title("Gray Sacle Image Histogram")
plt.xlabel("Gray levels")
plt.ylabel("No of Pixels")
```

```python
equalized_img = cv2.equalizeHist(gray_img)

# Display the Equalized Image
plt.imshow(equalized_img, cmap='gray')
plt.title("Equalized Gray Image")
plt.show()
```

```python
plt.hist(equalized_img.ravel(),256, [0,256])
plt.title("Equalized Image Histogram")
plt.xlabel("Gray levels")
plt.ylabel("No of Pixels")
```
