Adaptive-

Global Thresholding:
In global thresholding, a single threshold value is applied to the entire image.
This threshold value is manually determined or calculated based on certain criteria such as Otsu's method, which minimizes the intra-class variance.

Adaptive Thresholding:
In adaptive thresholding, the threshold value for each pixel is calculated based on the intensity values of its local neighborhood.
Instead of using a single threshold value for the entire image, adaptive thresholding calculates the threshold value independently for each pixel.

```python
img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
a1 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 199, 5)
plt.imshow(a1, cmap='gray')
```

```python
#This code performs thresholding on the input image img using Otsu's thresholding method, which automatically determines the optimal threshold value based on the image histogram.

a,b3=cv2.threshold(img,0, 255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print(a)
plt.imshow(cv2.cvtColor(b3, cv2.COLOR_BGR2RGB))# or plt.imshow(b3, cmap='gray')
```
